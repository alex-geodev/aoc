```python
#find locations of s
start_point = [(idx,item.index('^')) for idx,item in enumerate(data) if '^' in item][0]
```

```python
arr = []
for line in data:
    new_line = [letter for letter in line]
    arr.append(new_line)
map_arr = np.array(arr)
```

# Part 1

```python
def obstruction(point):
    return True if data[point[0]][point[1]] == '#' else False
def move(point:tuple, direction: str = None):
    
    if direction=='^':
        north = (point[0]-1,point[1])
        if obstruction(north):
            return ((point[0],point[1]+1), '>')
        else:
            return (north,direction)
    if direction=='>':
        east = (point[0],point[1]+1)
        if obstruction(east):
            return ((point[0]+1,point[1]),'v')
        else:
            return (east,direction)
    if direction=='v':
        south = (point[0]+1,point[1])
        if obstruction(south):
            return ((point[0],point[1]-1),'<')
        else:
            return (south,direction)    
    if direction=='<':
        west = (point[0],point[1]-1)
        if obstruction(west):
            return ((point[0]-1,point[1]),'^')
        else:
            return (west,direction)   
    
```

```python
in_bounds = True
point = start_point
direction='^'
width = len(data[0])
height = len(data)
walking_loop = map_arr.copy()
while in_bounds:
   walking_loop[point]= 1
   try:
      point,direction= move(point,direction)
      walking_loop[point]= 1
   except:
      in_bounds = False
```

# Part 2

```python
def obstruction(point, arr):
    return True if arr[point] == '#' else False

def move(point:tuple, arr: np.array, direction: str = None):

    if direction=='^':
        north = (point[0]-1,point[1])
        if obstruction(north, arr):
            #try turning to make sure not boxing in and reversing path
            east = (point[0],point[1]+1)
            if obstruction(east):
              return ((200,200),'v')
            return (east, '>')
        else:
            #print("moving north")
            return (north,direction)
    if direction=='>':
        east = (point[0],point[1]+1)
        if obstruction(east,arr):
            south = (point[0]+1,point[1])
            if obstruction(south):
              return ((200,200),'v')
            return (south,'v')
        else:
            #print("moving east")
            return (east,direction)
    if direction=='v':
        south = (point[0]+1,point[1])
        if obstruction(south,arr):
            west = (point[0],point[1]-1)
            if obstruction(west):
              return ((200,200),'v')
            return (west,'<')
        else:
            #print("moving south")
            return (south,direction)
    if direction=='<':
        west = (point[0],point[1]-1)
        if obstruction(west,arr):
            north = (point[0]-1,point[1])
            if obstruction(north):
              return ((200,200),'v')
            return (north,'^')
        else:
            #print("moving west")
            return (west,direction)
```

```python
rows,cols  = np.where(walking_loop=='1')
points_to_block = list(zip(rows,cols))
```

```python
def out_of_bounds(point):
  if point[0] >=10 or point[0]<0 or point[1]>=10 or point[1]<0:
    return True
  else:
    return False
```

```python
width = height = map_arr.shape[0]
loops = 0
for x in points_to_block:
  point = start_point
  direction='^'
  moves=0
  ob_arr = map_arr.copy()
  ob_arr[x] = '#'
  in_bounds = True
  points_traversed = []
  moves = 0 
  while in_bounds:
    try:
      ob_arr[point] = 1
      #if out_of_bounds(point):
      # in_bounds=False
      point,direction= move(point, ob_arr,direction)
      if point == (200,200):
        in_bounds=False
      moves+=1
      if moves>30000:
        loops+=1
        in_bounds=False
    except:
        in_bounds=False
  
loops
```
