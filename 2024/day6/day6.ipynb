```python
#find locations of s
start_point = [(idx,item.index('^')) for idx,item in enumerate(data) if '^' in item][0]
```

```python
arr = []
for line in data:
    new_line = [letter for letter in line]
    arr.append(new_line)
map_arr = np.array(arr)
```

# Part 1

```python
def obstruction(point):
    return True if data[point[0]][point[1]] == '#' else False
def move(point:tuple, direction: str = None):
    
    if direction=='^':
        north = (point[0]-1,point[1])
        if obstruction(north):
            print(f"hit ob: moving east - {data[north[0]][north[1]]}")
            return ((point[0],point[1]+1), '>')
        else:
            print("moving north")
            return (north,direction)
    if direction=='>':
        east = (point[0],point[1]+1)
        if obstruction(east):
            print(f"hit ob: moving south - {data[east[0]][east[1]]}")
            return ((point[0]+1,point[1]),'v')
        else:
            print("moving east")
            return (east,direction)
    if direction=='v':
        south = (point[0]+1,point[1])
        if obstruction(south):
            print(f"hit ob: moving west - {data[south[0]][south[1]]}")
            return ((point[0],point[1]-1),'<')
        else:
            print("moving south")
            return (south,direction)    
    if direction=='<':
        west = (point[0],point[1]-1)
        if obstruction(west):
            print(f"hit ob: moving north - {data[west[0]][west[1]]}")
            return ((point[0]-1,point[1]),'^')
        else:
            print("moving west")
            return (west,direction)   
    
```

```python
in_bounds = True
point = start_point
direction='^'
width = len(data[0])
height = len(data)
i=0


while in_bounds:
   map_arr[point]= 1
   try:
      point,direction= move(point,direction)
      map_arr[point]= 1
   except:
      in_bounds = False
```
